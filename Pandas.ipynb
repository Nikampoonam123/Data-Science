## Pandas filtering and Groupby

```python
df.head()
```

```python
#Extract record from gender = female
df['gender'] == 'Female'
```

```python
#Masking
df[df['gender']=='Female']
```

```python
df[df['gender'] == 'female']
```

```python
#Gender is female and salary is greater than 150000
print(df['gender'] == 'Female')
print(df['salary'] > 150000)
```

```python
df[(df['gender'] == 'Female') & (df['salary'] > 150000)]  
```

```python
df[(df['gender'] == "Female") | (df['salary'] > 150000)]
```

```python
#What is the average salary of rank prof?
```

```python
df[df['rank'] == 'Prof']
```

```python
df['salary'].mean()  #average of mentioned column
```

```python
#Average of prof's salary
df[df['rank'] == 'Prof']['salary'].mean()
```

```python
#Average salary of prof rank and gender is Male
df[(df['rank'] == "Prof") & (df['gender'] == 'Female')]['salary'].mean()
```

```python
#Average salary for each value in rank column
print(df[df['rank'] == 'Prof']['salary'].mean())
print(df[df['rank'] == 'AssocProf']['salary'].mean())
print(df[df['rank'] == 'AsstProf']['salary'].mean())
```

### Grouby aggregate method

```python
df.groupby('rank')['salary'].mean()  #easy to perform more operations like above ex.
```

```python
df.groupby('rank')['salary'].agg({'mean','median','min','max'}) #use aggregate to find many values of descriptive statistics
```

```python
df[df['salary'] == 186960]
```

```python
df[df['rank'] == 'Prof']['salary'].max()  #maximum salary for prof.
```
