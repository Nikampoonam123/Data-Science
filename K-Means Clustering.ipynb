# Elbow graph 

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
# Here i means we check different k values
```

```python
wcss 
# WCSS value for all K values from 1 to 10
```

```python
plt.plot(range(1,11),wcss)
plt.axvline(x = 5, linestyle = '--', color = 'red')
plt.xlabel('K-Values')
plt.ylabel('WCSS')
plt.title('Elbow Graph')
```

# Final Model

```python
km = KMeans(n_clusters=5)  # 5 clusters are formed
ypred = km.fit_predict(x)
ypred
```

```python
#Visualize the clusters
```

```python
km.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,1], c = ypred, cmap='rainbow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], c = 'yellow',s = 200)
```
