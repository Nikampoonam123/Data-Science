# Model building

```python
#Define x and y
x = df.iloc[:,:-1]
y = df['price']
```

```python
# Split the data
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
#Model building
lr = LinearRegression()
lr.fit(xtrain,ytrain)
ypred = lr.predict(xtest)
```

```python
# intercept
lr.intercept_
```

```python
# slope
lr.coef_
```

```python
# Accuracy
lr.score(xtrain,ytrain), lr.score(xtest,ytest)
# We see overfitting so we use regularization to deal with overfitting
```

# Lasso

```python
l1 = Lasso() # lassso regularization

l1.fit(xtrain,ytrain) # Add some error in training data

l1.score(xtrain,ytrain), l1.score(xtest,ytest)
```

```python
l1.coef_
# Slope of all columns
```

# Hyperparameter Tuning

```python
# Hyperparameter-> learning rate
# We have only one hyperparameter so we use for loop to get optimal alpha value
for i in range(100,200):
    l1 = Lasso(alpha=i)
    l1.fit(xtrain,ytrain)
    print(f"Alpha: {i} Train : {l1.score(xtrain,ytrain)} Test: {l1.score(xtest,ytest)}")
```

```python
# see diffent values and see where the train accuracy and test accuracy is similar. select those value
l1 = Lasso(alpha=175)
l1.fit(xtrain,ytrain)
l1.score(xtrain,ytrain), l1.score(xtest,ytest)
```

```python
l1.coef_
# Slope values and where the slope is very very less this slope values convets into zero
```

# Ridge 

```python
l2 = Ridge()

l2.fit(xtrain,ytrain)
l2.score(xtrain,ytrain), l2.score(xtest,ytest)
```

```python
for i in range(1,100):
    l2 = Ridge(alpha=i)
    l2.fit(xtrain,ytrain)
    print(f"Alpha: {i} Train: {l2.score(xtrain,ytrain)} Test: {l2.score(xtest,ytest)}")
```

```python
l2 = Ridge(alpha=7) # Select optimal alpha
l2.fit(xtrain,ytrain)
l2.score(xtrain,ytrain),l2.score(xtest,ytest)
```

```python
l2.coef_
# All slope but less slope not convert into 0
```
