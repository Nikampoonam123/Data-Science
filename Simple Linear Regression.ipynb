# Simple linear regression

- Ex. In order to investigatethe feasibility of starting a sunday edition for a largge matrapolitan newspaper, information was obtained from a sample 34 newspapers concerning their daily and sunday circulations(in thousands)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.dpi'] = 150
plt.rcParams['figure.figsize'] = (10,5)
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#Read data
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/NewspaperData.csv')
df
```

```python
#Select relevant columns
df = df[['daily','sunday']]
df
```

```python
#Correlation analysis
sns.scatterplot(x = df['daily'],y = df['sunday'])
```

```python
df.corr()  #correlation coefficient 
```

```python
sns.heatmap(df.corr(),annot=True)  # see to value in easy way
```

```python
#Import statsmodels
import statsmodels.formula.api as smf
```

```python
model = smf.ols('sunday~daily',data=df).fit()
model
```

```python
#Model parameters (slope, intercept)
model.params
```

```python
#line equation
daily_sales = 850

sunday_sales = 1.33*daily_sales+13.83
sunday_sales
```

```python
#Print p value
model.pvalues
```

```python
#r2 value
model.rsquared, model.rsquared_adj
```

## Prediction for new data

```python
df.head()
```

```python
test = pd.DataFrame({
    'daily' : [100,152,850,654,759]
})
test
```

```python
model.predict(test)
```
